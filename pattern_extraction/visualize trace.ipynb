# Accumulate canton trace matrix for daily trace

Split and read all daily traces and then accumulate cooresponding value. 

```python
trace=user_dtrace.reset_index()
trace.head()
```

```python
trace.loc[0]['state']
```

```python
trace.loc[0]['state'].split('->')
```

```python
lst_trace=[]
for s in trace['state']:
    places=s.split('->')
    lst_trace.append(places)
#lst_trace
```

```python
len(lst_trace)
```

```python
def obtain_canton(c):
    canton = canton_short[c]
    return canton

c='Bern - Berne'
obtain_canton(c)
```

```python
def accumulate_ctrace(c1,c2):
    i1=obtain_canton(c1)
    i2=obtain_canton(c2)
    canton_trace.loc[i1,i2]+=1
```

```python
for i in range(len(lst_trace)):
    for n in range(1,len(lst_trace[i])):
        accumulate_ctrace(lst_trace[i][n-1],lst_trace[i][n])
```

```python
def geo_convert_dec(degree,minute,direction):
    if direction == 'S' or direction == 'W':
        return -float(degree+minute/60.0)
    return float(degree+minute/60.0)

def match(i,la_d,la_m,d1,lo_d,lo_m,d2):
    c_coordinate.loc[i,'lat']= geo_convert_dec(la_d,la_m,d1)
    c_coordinate.loc[i,'lon']= geo_convert_dec(lo_d,lo_m,d2)
```

```python
map = folium.Map(location=[46.82244,8.22410], zoom_start=8)
map.choropleth(
               key_on='feature.id', 
               geo_path=CH_GEO_JSON, 
               topojson='objects.cantons', 
               fill_color='YlOrRd'
               )
points = []
for index in range(len(c_coordinate)):
    points.append(tuple([c_coordinate.loc[index]['lat'], c_coordinate.loc[index]['lon']]))

#add a markers
for each in points:  
    folium.RegularPolygonMarker(each,fill_color='#43d9de',radius=5).add_to(map)
map
```

```python
#fadd lines
lines=[]
n_len=len(canton_ids)
for i in range(n_len):
    for j in range(n_len):
        popular = canton_trace[canton_ids[i]][canton_ids[j]]+canton_trace[canton_ids[j]][canton_ids[i]]
        if(popular!=0):
            lines=tuple([points[i],points[j]])
            folium.PolyLine(lines, color="red", weight=0.3*np.sqrt(popular), opacity=0.4).add_to(map)
map
```
